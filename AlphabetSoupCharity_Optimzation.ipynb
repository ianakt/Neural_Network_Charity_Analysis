```python


# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler,OneHotEncoder
import pandas as pd
import tensorflow as tf
#  Import and read the charity_data.csv.
import pandas as pd 
application_df = pd.read_csv("charity_data.csv")
application_df.head()
```

```python


application_df = application_df.drop(columns=["EIN","NAME"])
application_df.head()


```

```python
count = [len(application_df[application_df.columns[x]].value_counts()) for x in range(len(application_df.columns))]
count
```

```python
application_df['APPLICATION_TYPE'].value_counts()
application_types_to_replace = ['T9','T13','T12','T2','T25','T14','T29','T15','T17']
```

```python


# Choose a cutoff value and create a list of application types to be replaced
# use the variable name `application_types_to_replace`
#  YOUR CODE GOES HERE

# Replace in dataframe
for app in application_types_to_replace:
    application_df['APPLICATION_TYPE'] = application_df['APPLICATION_TYPE'].replace(app,"Other")

# Check to make sure binning was successful
application_df['APPLICATION_TYPE'].value_counts()

```

```python


# Look at CLASSIFICATION value counts for binning
#  YOUR CODE GOES HERE

application_df['CLASSIFICATION'].value_counts()

```

```python
# You may find it helpful to look at CLASSIFICATION value counts >1
application_df['CLASSIFICATION'].value_counts().loc[lambda x : x>1] 
```

```python
classifications_to_replace = application_df['CLASSIFICATION'].value_counts()
```

```python


# Choose a cutoff value and create a list of classifications to be replaced
# use the variable name `classifications_to_replace`
#  YOUR CODE GOES HERE

# Replace in dataframe
for cls in classifications_to_replace:
    application_df['CLASSIFICATION'] = application_df['CLASSIFICATION'].replace(cls,"Other")
    
# Check to make sure binning was successful
application_df['CLASSIFICATION'].value_counts()

```

```python
replace_ask_amt = application_df.ASK_AMT.value_counts()[application_df.ASK_AMT.value_counts() > (0.05*application_df.ASK_AMT.value_counts().sum())].index.to_list()
for org in replace_ask_amt:
    application_df.ASK_AMT = application_df.ASK_AMT.replace(org, 'Other')
```

```python
application_df.info()
```

```python
new_application_df = application_df.drop('IS_SUCCESSFUL', axis=1)
```

```python


# Convert categorical data to numeric with `pd.get_dummies`
#  YOUR CODE GOES HERE
x = pd.get_dummies(new_application_df)

x.head(5)

```

```python


# Split our preprocessed data into our features and target arrays
#  YOUR CODE GOES HERE

y = application_df['IS_SUCCESSFUL'].values
X = x.values

```

```python
# Split the preprocessed data into a training and testing dataset

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=55)
```

```python


# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)

```

```python
number_input_features = len(X_train[0])
number_input_features
```

```python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.
#  YOUR CODE GOES HERE

nn = tf.keras.models.Sequential()

# First hidden layer
#  YOUR CODE GOES HERE

nn.add(tf.keras.layers.Dense(units=80, activation="relu", input_dim=number_input_features))

# Second hidden layer
#  YOUR CODE GOES HERE

nn.add(tf.keras.layers.Dense(units=20, activation="tanh"))

nn.add(tf.keras.layers.Dense(units=33, activation='elu'))
nn.add(tf.keras.layers.Dense(units=44, activation='exponential'))
nn.add(tf.keras.layers.Dense(units=22, activation='gelu'))
nn.add(tf.keras.layers.Dense(units=44, activation='hard_sigmoid'))
nn.add(tf.keras.layers.Dense(units=44, activation='linear'))
nn.add(tf.keras.layers.Dense(units=44, activation='selu'))
nn.add(tf.keras.layers.Dense(units=39, activation='softmax'))
nn.add(tf.keras.layers.Dense(units=44, activation='softplus'))
nn.add(tf.keras.layers.Dense(units=44, activation='swish'))
nn.add(tf.keras.layers.Dense(units=44, activation='tanh'))


# Output layer
#  YOUR CODE GOES HERE

nn.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))

# Check the structure of the model
nn.summary()
```

```python
#
nn.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
```

```python


# Train the model
#  YOUR CODE GOES HERE

fit_model = nn.fit(X_train_scaled, y_train, epochs=10)

```

```python


# Evaluate the model using the test data
model_loss, model_accuracy = nn.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")

```
